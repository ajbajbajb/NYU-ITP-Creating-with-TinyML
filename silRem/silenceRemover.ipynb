```python
from pydub import AudioSegment
import glob

dirs = glob.glob('gunshots_store/*/')
print(dirs)
audioFiles = []
for dir in dirs:
    # and here i learn the beauty of extend over append...
    # extend doesn't put arrays inside arrays, it just adds to this one
    audioFiles.extend(glob.glob(dir + '*.wav'))
print(audioFiles)
```

# To Test On ALL FILES IN A DIRECTORY

```python
def detect_leading_silence(sound, silence_threshold=-15.0, chunk_size=10):
    '''
    sound is a pydub.AudioSegment
    silence_threshold in dB
    chunk_size in ms

    iterate over chunks until you find the first one with sound
    '''
    trim_ms = 0  # ms

    assert chunk_size > 0  # to avoid infinite loop
    while sound[trim_ms:trim_ms+chunk_size].dBFS < silence_threshold and trim_ms < len(sound):
        trim_ms += chunk_size

    return trim_ms


for soundFile in audioFiles:
    sound = AudioSegment.from_file(soundFile, format="wav")
    start_trim = detect_leading_silence(sound)
    end_trim = detect_leading_silence(sound.reverse())
    duration = len(sound)
    trimmed_sound = sound[start_trim:duration-200]
    trimmed_sound.export(('gunshots_silenced/SR_' + soundFile.replace('gunshots_store/', '') + '_silrem.wav'), format="wav")


```
